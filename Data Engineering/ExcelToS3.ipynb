# Parsing list of forms for each visit

```python
forms_df = pd.read_excel(r'C:\Users\siuniaev\Documents\@projects\SBAT\MK8807-009_GDMS_Intake.xlsm', sheet_name='10. T&E WS', skiprows=7)
forms_df
```

### drop everything below forms

```python
forms = forms_df[forms_df.columns[1]]
k = 0
for index, value in forms.items():
    #print(f"Index : {index}, Value : {type(value)}")
    if (not isinstance(value, str)):
        k = index - 1
        
#forms_df.drop(forms.index[-1:k-1:-1])
forms_df = forms_df.drop(forms.index[-1:k-1:-1])
```

### save form refname as Series

```python
form_refnames = forms_df[forms_df.columns[1]]
form_refnames
```

```python
forms_df = forms_df.filter(like='Unnamed', axis=1)
forms_df
```

```python
df.loc[df.index[2]]
```

```python
forms_df.columns = df.loc[df.index[1]]
forms_df
```

### trying to aggregate the list of forms

```python
forms_df[form_refnames.name] = form_refnames
visit1 = forms_df[forms_df.columns[[0,-1]]]
visit1 = visit1.applymap(str)

visit1 = visit1[visit1[visit1.columns[0]].str.match('^([0-9])')]
counts = visit1[visit1.columns[0]]
forms = visit1[visit1.columns[1]]
result=[1, 2]
for form, count in zip(forms.values, counts.values):
    result[int(count)] = form
result
```
